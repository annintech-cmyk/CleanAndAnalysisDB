```python
def filter_major_category():
    major_category = all_ages["Major_category"].dropna()
    major_category_list = []
    for category in major_category:
        if category not in major_category_list:
            major_category_list.append(category)
    return major_category_list
print(filter_major_category())
```

```python
def filter_major_category(ref):
    return ref.dropna().sort_values().unique().tolist()
print(filter_major_category(all_ages["Major_category"]))
```

```python
major_category_from_all_ages = all_ages["Major_category"]
major_category_from_all_ages_list = filter_major_category(major_category_from_all_ages)
print(major_category_from_all_ages_list)
print(major_category_from_all_ages_list.count("Physical Sciences"))
print(len(major_category_from_all_ages_list))
```

```python
major_category_from_recent_grads = recent_grads["Major_category"]
major_category_from_recent_grads_list = filter_major_category(major_category_from_recent_grads)
print(major_category_from_recent_grads_list)
print(major_category_from_recent_grads_list.count())
```

```python
def calculate_major_cat_totals(df):
    cats  = df['Major_category'].unique()
    counts_dictionary = dict()

    for c in cats:
        major_df = df[df['Major_category'] == c]
        total = int(major_df["Total"].sum())
        counts_dictionary[c] = total
    return counts_dictionary
```

```python
def calculate_total(df):
    return (df.groupby("Major_category")["Total"].sum().sort_values().astype(int).to_dict())

print(calculate_total(all_ages))
```

```python
aa_cat_total = calculate_major_cat_totals(all_ages)
print(aa_cat_total)
```

```python
rg_cat_total = calculate_major_cat_totals(recent_grads)
print(rg_cat_total)
print(rg_cat_total)
```
