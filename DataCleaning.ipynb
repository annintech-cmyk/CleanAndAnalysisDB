```python
df["brand_name"] = df["brand_name"].apply(lambda x: str(x).capitalize())
print(df.head())
```

```python
df["brand_name"] = df["brand_name"].apply(lambda x: x[0:3]+"-"+x[3:])
```

```python
df["brand_name"] = df["brand_name"].str.strip().str.lower().replace(["n/a-", "nan", "nn", "n/a"], "").str.capitalize()
print(df["brand_name"] )
```

```python
df[["brand", "name"]] = df["brand_name"].str.split('-', n=1, expand=True)
df["brand_name"] = df["brand_name"].str.replace("-","")

print(df.head())
```

```python
df["brand_name"] = df["brand_name"].str.replace("Nan","Dermalux")
df["brand_name"]
```

```python
df = df[df["brand_name"].str.contains(r'[a-zA-Z0-1]', na=False)]
print(df)
```

```python
for x in df.index:
    if df.loc[x, "name"] == " ":
        df.loc[x, "name"] = "XX"
print(df)
```
