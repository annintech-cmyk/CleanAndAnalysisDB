```python
# ----- While loop tutorial ----

limit = input("Enter the range?")
limit_ref = int(limit)
print("Start printing"+ str(limit_ref))
n = 0
while n < limit_ref:
    n = n + 1
    print(n)


print("End printing"+ str(limit_ref))


```

```python
def get_ref(*args, **kwargs):
    print("Enter args:", args)

    for n, arg in enumerate(args, start=1):
        print(f"Arg {n}: {arg}")

    print("Enter kwargs:", kwargs)

    for key, value in kwargs.items():
        print(f"Key {key}: Value {value}")


get_ref(1, 2, "Ref arg", age = 2, name = "Ann")
```

```python
def ref_func(*args, **kwargs):
    print("Enter args:", args)
    for n, args in enumerate(args, start=1):
        print(f"Arg {n}: {args}")
    for key, value in kwargs.items():
        print(f"Key {key}: Value {value}")

ref_func("Ref arg", 3,44,55,66, age = 2, name = "Ann")
```

```python
def magic_age():
    age = input("Enter your age:")
    age_magic = int(age)
    if int(age) < 10:
        print("Child!")
        age_magic = 5
    elif int(age) < 20:
        print("Teenage!")
        age_magic = 10
    elif int(age) < 40:
        print("Adult!")
        age_magic = 30
    else:
        print("Awsome!")
        age_magic = 50
    return age_magic

if magic_age() == 50:
    print("Welcome!")


```
