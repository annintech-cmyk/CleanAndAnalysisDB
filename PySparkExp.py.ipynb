```python
def create_spark_session(name = "exp"):
    return  SparkSession.builder.appName(name).getOrCreate()
```

```python
# function to create RDD
def create_RDD(sc_obj, data_ref):
    return sc_obj.parallelize(data_ref)
```

```python
# function to convert RDD to dataframe
def RDD_to_df(rd_df_ref,schema):
  # converting RDD to dataframe using toDF()
  # in which we are passing schema of df
  return  rd_df_ref.toDF(schema)
```

```python
if __name__ == "__main__":

  input_data = [("Alex",122000,89600,12238),
          ("Emma",454000,380000,67985),
          ("Erina",115000,102000,13933),
          ("Angel",147000,111000,15306),
          ("Jerome",153000,124000,5259)]

  # calling function to create SparkSession
  spark = create_spark_session("Exp_spark")

  # creating spark context object
  sc = spark.sparkContext

  # calling function to create RDD
  rd_df = create_RDD(sc,input_data)

  schema_lst = ["Name","Assets","SpendingCapacity","Salary"]

  # calling function to convert RDD to dataframe
  converted_df = RDD_to_df(rd_df,schema_lst)

  # visualizing the schema and dataframe
  converted_df.printSchema()
  converted_df.show()
```

```python
#Join Data sets
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("N").getOrCreate()

d1 = [("1", "sravan", "company 1"),
      ("2", "ojaswi", "company 1"),
      ("3", "rohith", "company 2"),
      ("4", "sridevi", "company 1"),
      ("5", "bobby", "company 1")]
cols1 = ['id', 'name', 'company']
df1 = spark.createDataFrame(d1, cols1)

d2 = [("1", "45000", "IT"),
      ("2", "145000", "Manager"),
      ("6", "45000", "HR"),
      ("5", "34000", "Sales")]
cols2 = ['id', 'salary', 'department']

df2 = spark.createDataFrame(d2, cols2)
print("DF1")
df1.show()
print("DF2")
df2.show()
print("inner")
df1.join(df2, df1.id == df2.id, "inner").show()
print("left")
df1.join(df2, df1.id == df2.id, "left").show()
print("right")
df1.join(df2, df1.id == df2.id, "right").show()
print("right_outer")
df1.join(df2, df1.id == df2.id, "right_outer").show()
print("full")
df1.join(df2, df1.id == df2.id, "full").show()
print("left_semi")
df1.join(df2, df1.id == df2.id, "left_semi").show()
print("left_anti")
df1.join(df2, df1.id == df2.id, "left_anti").show()
print("cross")
df1.join(df2, df1.id == df2.id, "cross").show()

df1.join(df2, df1.id == df2.id, "cross").show()
```

```python
# import necessary libraries
from pyspark.sql import SparkSession

# Create Spark Session
sparkSession = SparkSession.builder.appName('g1').getOrCreate()

# Create Spark DataFrame
df_pyspark = sparkSession.read.csv(
    'file/Employee_Table.csv',
    header=True,
    inferSchema=True
)

# Print Schema
df_pyspark.printSchema()
print("Before Data Cleaning")
# Print Dataframe
df_pyspark.show()
```

```python
#--------------------------------------------- Cleaning Data with PySpark Python ------------------------------------------------------
# Print Schema
df_pyspark.printSchema()
print("Before Data Cleaning")
# Print Dataframe
df_pyspark.show()

ref_Department = 'Department'
# Fill Null values inside Department column with the word 'Generalist'
df_pyspark = df_pyspark.na.fill('DEF', subset=ref_Department)
df_pyspark.printSchema()
# Assumed Null Value means Employee joined during Company Founding i.e. 2010
df_pyspark = df_pyspark.na.fill(2010, subset=['Joining Year'])

# Replace Information Tech with IT
df_pyspark = df_pyspark.replace(
    {'Information Tech': 'IT'}, subset=ref_Department)

# Remove Outlier -- We assume 59 to be maximum working age in the company
df_pyspark = df_pyspark.filter('Age<60')

# Remove Rows not containing First as well as Last Name
df_pyspark = df_pyspark.filter(
    df_pyspark['Last Name'].isNotNull() & df_pyspark['First Name'].isNotNull())
print("After Data Cleaning")
df_pyspark.drop('Department').show()
df_pyspark.show()
```
