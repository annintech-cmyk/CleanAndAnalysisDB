```python
import tensorflow as tf

# Create a scalar tensor
scalar = tf.constant(7)

# Create a vector tensor
vector = tf.constant([1, 2, 3, 4])

# Create a matrix tensor
matrix = tf.constant([[1, 2], [3, 4]])

# Create a 3D tensor
tensor_3d = tf.constant([[[1], [2]], [[3], [4]]])

print(scalar)
print(vector)
print(matrix)
print(tensor_3d)
```
